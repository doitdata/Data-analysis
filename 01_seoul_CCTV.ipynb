 ## 1-2 엑셀 파일 읽기 - 서울시 인구현황

```python
# 엑셀 파일도 데이터프레임으로 불러오고 싶으면 pd.read_excel() 사용하고, 인코딩은 utf-8이나 cp949로 해주면 된다. 
pop_seoul = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
# 상위 5개만 봅니다. 
pop_seoul.head()
```

```python
# read_csv(filepath, index_col=None, header='infer', names=None, ...)
# filepath는 파일의 위치 경로
# index_col은 index를 지정해주는 것으로 0이나 integer(위치)를 쓰거나 직접 변수 이름을 지정할 수 있다. 
# header는 첫행을 지정하는 것을 의미
# usecols는 불러올 칼럼의 인덱스 번호나 이름을 지정하는 것

pop_seoul=pd.read_excel('./data/01. population_in_Seoul.xls',
                       header = 2,
                       usecols='B,D,G,J,N',
                       encoding='utf-8')
pop_seoul.head()
```

```python
# 이름 재설정 rename(columns = {원래컬럼의 위치: '바뀔이름'})
# inplace=True => 변수의 내용 갱신하라는 의미

pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                         pop_seoul.columns[1]:'인구수',
                         pop_seoul.columns[2]:'한국인',
                         pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자'}, inplace=True)
                          
pop_seoul.head()
```

```python
# polyfit => 데이터를 대표하는 직선 하나 그리기. 
fp1 =np.polyfit(data_result['인구수'], data_result['소계'],1)
fp1
```

참고주소: https://076923.github.io/posts/Python-numpy-16/

```python
# poly1d는 다항식을 생성 함수로 입력배열을 항으로 간주하여 다항식을 생성
# np.poly1d(입력 배열, 근 설정, 변수명)으로 다항식을 생성
f1 = np.poly1d(fp1) #y축
# linespace 함수는 1차원 배열 만들기나 그래프 그리기에서 수평축의 간격 만들기에 사용
# lnespace(start, stop, num) 시자값, 끝값, 간격으로 보면됨. 
fx = np.linspace(100000, 700000, 100) #x축
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
# ls = line_style , lw = line_weight
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

## 1-7 조금 더 설득력 있는 자료 만들기

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
# abs 는 절대값
data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending =False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],
           c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
